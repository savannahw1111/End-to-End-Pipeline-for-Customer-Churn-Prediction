```python
# Importing customer info
import pandas as pd
customer_info = pd.read_csv("ecommerce_customer_features.csv")
display(customer_info.head())

# Importing target info
import pandas as pd
targets = pd.read_csv("ecommerce_customer_targets.csv")
display(targets.head())

# Merging the datasets with an inner join
df = customer_info.merge(targets, on="Customer_ID", how="inner")

# Viewing the head and tail
df.head()
df.tail()

```

```python

# Data shape and columns
print("SHAPE:", df.shape)
print("\nCOLUMNS:")
print(df.columns)

# Data info
print("\nINFO:")
df.info()

# Data summary statistics 
print("\nDESCRIBE:")
display(df.describe())

# Looking for missing values
print("\nMISSING VALUES (%):")
display((df.isnull().sum() / len(df)) * 100)

# Looking for duplicates
print("\nDUPLICATE ROWS:", df.duplicated().sum())
print("DUPLICATE Customer_ID:", df["Customer_ID"].duplicated().sum())

```

```python
# Connecting SQL database
!pip install duckdb
import duckdb
con = duckdb.connect()
con.register("raw_customers", df)

# Executing SQL query
clean_df = con.execute("""
-- Table
WITH clean_df AS (
  SELECT
    Customer_ID,
    account_age_months,
    avg_order_value,
    total_orders,
    days_since_last_purchase,
    discount_usage_rate,
    return_rate,
    customer_support_tickets,
    browsing_frequency_per_week,
    cart_abandonment_rate,
    product_review_score_avg,
    engagement_score,
    satisfaction_score,
    price_sensitivity_index,

    -- Encode categoricals
    CASE WHEN LOWER(TRIM(loyalty_member)) IN ('yes','y','true','1','member') THEN 1 ELSE 0 END AS loyalty_flag,

    CASE WHEN LOWER(TRIM(churned)) IN ('yes','y','true','1','churned') THEN 1 ELSE 0 END AS churn_label,

    -- Feature Engineering
    (avg_order_value * total_orders) AS est_lifetime_revenue,

    CASE WHEN days_since_last_purchase >= 60 THEN 1 ELSE 0 END AS dormant_60d,

    CASE
      WHEN account_age_months > 0 THEN total_orders * 1.0 / account_age_months
      ELSE NULL
    END AS orders_per_month
  FROM raw_customers
  WHERE Customer_ID IS NOT NULL
)
SELECT *
FROM clean_df
WHERE
  churn_label IS NOT NULL
  AND account_age_months >= 0
  AND avg_order_value >= 0
  AND total_orders >= 0
  AND days_since_last_purchase >= 0
  AND discount_usage_rate BETWEEN 0 AND 1
  AND return_rate BETWEEN 0 AND 1
  AND cart_abandonment_rate BETWEEN 0 AND 1
  AND product_review_score_avg BETWEEN 0 AND 5
  AND engagement_score BETWEEN 0 AND 10
  AND satisfaction_score BETWEEN 0 AND 10;
""").fetchdf()

clean_df.head()
```
