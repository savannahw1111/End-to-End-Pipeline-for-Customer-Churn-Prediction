# IMPORTING DATA

```python
# Importing customer info
# From https://www.kaggle.com/datasets/vishardmehta/e-commerce-customer-churn-prediction-dataset 
import pandas as pd
customer_info = pd.read_csv("C:/Users/savan/Documents/Projects/End-to-End-Pipeline-for-Customer-Churn-Prediction/data/ecommerce_customer_features.csv")
display(customer_info.head())

# Importing target info
import pandas as pd
targets = pd.read_csv("C:/Users/savan/Documents/Projects/End-to-End-Pipeline-for-Customer-Churn-Prediction/data/ecommerce_customer_targets.csv")
display(targets.head())

# Merging the datasets with an inner join
df = customer_info.merge(targets, on="Customer_ID", how="inner")

# Viewing the head and tail
df.head()
df.tail()

```

```python
# CLEANING/ENCODING
```

```python
# Connecting SQL 
!pip install duckdb
import duckdb
con = duckdb.connect()
con.register("raw_customers", df)

# Executing SQL query and decoding strings to numbers
clean_encoded_df = con.execute("""
-- Table
WITH clean_encoded_df AS (
  SELECT *,

    -- Encode categoricals
    CASE WHEN LOWER(TRIM(loyalty_member)) IN ('yes','y','true','1','member') THEN 1 ELSE 0 END AS loyalty_flag,

    CASE WHEN LOWER(TRIM(churned)) IN ('yes','y','true','1','churned') THEN 1 ELSE 0 END AS churn_label,
    FROM raw_customers
    WHERE Customer_ID IS NOT NULL
)
SELECT *
FROM clean_encoded_df
WHERE
  churn_label IS NOT NULL
  AND account_age_months >= 0
  AND avg_order_value >= 0
  AND total_orders >= 0
  AND days_since_last_purchase >= 0
  AND discount_usage_rate BETWEEN 0 AND 1
  AND return_rate BETWEEN 0 AND 1
  AND cart_abandonment_rate BETWEEN 0 AND 1
  AND product_review_score_avg BETWEEN 0 AND 5
  AND engagement_score BETWEEN 0 AND 10
  AND satisfaction_score BETWEEN 0 AND 10;
""").fetchdf()

clean_encoded_df.head()
```

```python
# FEATURE ENGINEERING
```

```python

# Feature Engineering 
con.register("clean_encoded_df", clean_encoded_df)
processed_df = con.execute("""
    WITH engineered AS (
    SELECT *,
    
   -- Lifetime revenue
   (avg_order_value * total_orders) AS est_lifetime_revenue,
   
    -- Days since last purchase
    CASE WHEN days_since_last_purchase >= 60 THEN 1 ELSE 0 END AS dormant_60d,
    
    -- Orders per month
    CASE
      WHEN account_age_months > 0 THEN total_orders * 1.0 / account_age_months
      ELSE NULL
    END AS orders_per_month,
    
    -- Help tickets per order
    CASE 
      WHEN total_orders > 0 THEN customer_support_tickets * 1.0 / total_orders
      ELSE 0
    END AS tickets_per_order,
    
    -- Return volume
    (return_rate * total_orders) AS return_volume
    FROM clean_encoded_df
    )
SELECT *,
FROM engineered
""").fetchdf()

processed_df.head()

```

```python
# Dropping unnecessary (text) columns
final_df = processed_df.drop(columns=["loyalty_member", "churned", "Customer_ID"])

# Saving
processed_df.to_csv("../data/processed_df.csv", index=False)
final_df.to_csv("../data/final_df.csv", index=False)
```
